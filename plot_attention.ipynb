```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_vsn = np.load('f_vsn.npy')
h_vsn = np.load('h_vsn.npy')
sparse_weights = np.load('sparse_weights.npy')
```

```python
np.mean(np.mean(f_vsn, axis=0), axis=0)
```

```python
np.mean(np.mean(h_vsn, axis=0), axis=0)
```

```python
np.mean(sparse_weights, axis=0)
```

```python
plt.figure(figsize=(12, 8))
plt.barh(
    np.arange(f_vsn.shape[-1]),
    np.mean(np.mean(f_vsn, axis=0), axis=0),
    tick_label=['DAY_OF_WEEK', 'IS_WEEKEND', 'WEATHER'])
plt.plot()
```

```python
plt.figure(figsize=(12, 8))
plt.barh(
    np.arange(h_vsn.shape[-1]),
    np.mean(np.mean(h_vsn, axis=0), axis=0),
    tick_label=['DAY_OF_WEEK', 'IS_WEEKEND', 'WEATHER', 'ACTIVITY'])
plt.plot()
```

```python
plt.figure(figsize=(12, 8))
plt.barh(
    np.arange(sparse_weights.shape[-1]),
    np.mean(sparse_weights, axis=0),
    tick_label=['SEX', 'AGE_GROUP', 'INCOME'])
plt.plot()
```

```python
attn = np.load('attn.npy')
np.mean(np.mean(attn, axis=0), axis=0)
```
